```python
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names() 
```

```python
a=sns.load_dataset("tips")
a.head()
```

**stripplot**

```python
a=sns.load_dataset("tips")
sns.stripplot(x="sex",y="smoker", data=a,hue='smoker',jitter=True,dodge=True)
sns.set_style("ticks")
plt.title("smokers - sex")
plt.show
```

**Scatterplot**

```python
sns.scatterplot(x='total_bill',y='tip',marker='o',color='g',data=a, hue="sex")
sns.set(style="whitegrid")  # dark,white,darkgrid,whitegrid,ticks
plt.show
```

**lineplot**

```python
sns.lineplot(x='total_bill',y='tip',data=a)
plt.show
```

```python
sns.lineplot(x='total_bill',y='tip',marker='o',color='r',data=a)
plt.grid()
plt.legend('size',loc='upper left')
plt.show
```

```python
sns.lineplot(x='total_bill',y='tip',data=a,marker=".",ms=22,mec="blue",mfc="y",color="green") # ms=marker size mec marker edge size mec marker edge color 
plt.legend("Height",loc="lower right")
plt.title("Ticks")
sns.set_style('dark')
```

**swarmplot**

```python
s=sns.load_dataset("iris")
s.head()
sns.swarmplot(x='sepal_length',y='species',data=s)
sns.set_style('darkgrid')
```

**barplot**

```python
sns.barplot(x='sex',y='total_bill',palette='plasma',data=a)
```
